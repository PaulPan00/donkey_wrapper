```python
# class generater
class Gen:
    def __init__(self, n):
        self.n = n
        self.last = 0
    
    def __next__(self):
        return self.next()
    
    def next(self):
        if self.last == self.n:
            raise StopIteration()

        rv = self.last ** 2
        self.last += 1
        return rv

g = Gen(10000)

while 1:
    try:
        print(next(g))
    except StopIteration:
        break
```

```python
def gen(n):
    for i in range(n):
        # pausing execution
        yield i ** 2

g = gen(10000)

# for i in g:
#     print(i)
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
print(next(g))
```

```python
def another_gen():
    yield 1
    yield 5
    yield 8

gg = another_gen()

print(next(gg))
print(next(gg))
print(next(gg))
print(next(gg)) # won't yield any more
```
